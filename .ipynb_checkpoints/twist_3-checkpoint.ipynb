# East Africa Analysis of Violent Activities

## Overview

Violence activities, to some extent, have an impact on a country's economic stability. It is therefore very important to monitor and evaluate the level of violent activities and distractions for better preparation and control and to avoid them, where necessary, for a better economy.

In an attempt to understand the types and relations of violence in East Africa, we used The Armed Conflict Location & Event Data Project (ACLED) dataset for East African countries. ACLED collects, analyzes, and maps reported crises, including political violence and protest events around the world. The East Africa data has information from 1997 to date for the following countries: Kenya, Ethiopia, Somalia, South Sudan, Burundi, Mozambique, Uganda, Madagascar, Tanzania, Djibouti, Rwanda, Eritrea, Malawi, Mauritius, Comoras, Mayotte, Reunion, and Seychelles.

The main objective was to analyse the pattern of events or types of violence in East Africa over time. What are the levels of violent activity in Kenya and East Africa?

We cleaned the data by removing anything that is not a word or space. We also realised that the months were dominating some of the wordclouds and removed them from the text.

```python
#defining a function that will clean the data
def text_cleaner (text):
    """
    A function to clean the tweet text
    """
        
    try: 
        text = re.sub(r'[^\w\s]','',text) ## remove everything apart from words and space
        text = re.sub(r'_',' ',text) ## remove underscore
        text = re.sub(r'\n',' ',text) ## remove \n
         
    except TypeError: 
        pass 

    
    return text
```

```python
#cleaning the test and saving it into a new datafrane
df['clean_notes'] = df['notes'].apply(text_cleaner)
df= df[['event_date', 'year', 'event_type', 'sub_event_type', 'notes', 'clean_notes', 'iso3']]
df.head(5)
```

```python
# A list of months
months = ['January','February','March',
          'April','May','June','July',
          'August','september','October',
          'November','December']

# A fuction to remove the months on the notes text
def remove_months(text):
    
    """
    A function to remove month words
    """
    
    # Tokenize the text
    tokens = tokenizer.tokenize(text)
    tokens = [token.strip() for token in tokens]
    
    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in months]
    filtered_text = ' '.join(filtered_tokens)    
    
    return filtered_text
```

```python
#removing the months in the notes 
df['clean_notes'] = df['clean_notes'].apply(remove_months)
```

```python
#converting the notes in to string 
df['clean_notes'] = df['clean_notes'].astype(str)
```

```python
## Plot word cloud
all_words = ' '.join(twts for twts in df['clean_notes'])

text_cloud = wordcloud.WordCloud(height=300,width=500,random_state=10,max_font_size=110).generate(all_words) #background_color='white',

plt.figure(figsize=(10,8))
plt.title('Full Data Wordcloud')
plt.imshow(text_cloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#group the data by year and country
data = df.groupby(['year','iso3']).size()
data
```

```python
#reset the index of the data
data.reset_index()
data.reset_index(name='counts')
```
